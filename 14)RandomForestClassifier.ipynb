```python
X = df.drop("income",axis=1)
y = df["income"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
for i in [X_train, X_test]:
    i["workclass"] = i["workclass"].fillna(X_train["workclass"].mode()[0])
    i["occupation"] = i["occupation"].fillna(X_train["occupation"].mode()[0])
    i["native-country"] = i["native-country"].fillna(X_train["native-country"].mode()[0])
```

```python
#native-country de 41 farklı çeşit var bu yüzden mean encoding yapmak mantıklı
```

```python
y_train_binary = y_train.apply(lambda x: 1 if x.strip()== ">50K" else 0)
```

```python
y_train_binary
```

```python
y_train_binary.groupby(X_train["native-country"]).mean()
```

```python
target_means = y_train_binary.groupby(X_train["native-country"]).mean()
```

```python
X_train["native_country_encoded"] = X_train["native-country"].map(target_means)
X_train["native_country_encoded"] = X_train["native_country_encoded"].fillna(y_train_binary.mean())

X_test["native_country_encoded"] = X_test["native-country"].map(target_means)
X_test["native_country_encoded"] = X_test["native_country_encoded"].fillna(y_train_binary.mean())
```

```python
X_train = X_train.drop("native-country",axis=1)
X_test = X_test.drop("native-country",axis=1)
```

```python
categorical
```

```python
one_hot_categories = [
    'workclass',
     'education',
     'marital-status',
     'occupation',
     'relationship',
     'race',
     'sex']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
encoder = ColumnTransformer(
    transformers=[
        ("cat",OneHotEncoder(handle_unknown="ignore",sparse_output=False),one_hot_categories)
    ],remainder="passthrough"
)
```

```python
X_train_enc = encoder.fit_transform(X_train)
X_test_enc = encoder.transform(X_test)
```

```python
# this gives X_train as numpy array if you want to turn this into df you have to get column names
```

```python
X_train_enc
```

```python
# get all encoded column names
columns = encoder.get_feature_names_out()
```

```python
columns
```

```python
X_train = pd.DataFrame(X_train_enc, columns=columns, index=X_train.index)
X_test = pd.DataFrame(X_test_enc, columns=columns, index=X_test.index)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=10,random_state=0)
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
rfc = RandomForestClassifier(n_estimators=100,random_state=15)
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
feature_score = pd.Series(rfc.feature_importances_, index=X_train.columns).sort_values(ascending=False)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rfc = RandomForestClassifier()
```

```python
params = {
    "n_estimators": [100,200,500,1000],
    "max_depth": [5,8,10,None],
    "ccp_alpha": [0,1,2,5],
    "max_features": ["sqrt","log2",5,10,None],
    "min_samples_split": [2,8,15,20]   
}
```

```python
rsCV = RandomizedSearchCV(estimator=rfc,param_distributions=params, cv=3, n_jobs=-1)
rsCV.fit(X_train,y_train)
```

```python
y_pred = rsCV.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
