```python
df = pd.read_csv("/kaggle/input/insurance/insurance.csv")
```

```python
df.head(3)
```

```python
pd.get_dummies(df,columns =['sex','smoker','region',],drop_first = True)
```

```python
df = pd.get_dummies(df,columns =['sex','smoker','region',],drop_first = True)
```

```python
df.head(3)
```

```python
y = df[['charges']]
x = df.drop('charges',axis=1)
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[21,25,0,1,1,0,0,1]])
```

```python
df_hata = pd.DataFrame()
```

```python
df_hata['y'] = y
```

```python
y_tahmin = model.predict(x)
```

```python
df_hata['tahmin'] = y_tahmin
```

```python
df_hata  #goruldugu gibi tahmin ve gerçek degerler arasında hata payları mevcuttur bunu net şekilde görebilmek için bu şekilde bir df olusturduk.
```

```python
df_hata['error'] = y - y_tahmin
```

```python
df_hata   # + ve - degerde hatalar oldugu için bunları MSE( Min Squared Error) yapmamız gerek
```

```python
df_hata['squared error'] = df_hata['error']**2
```

```python
df_hata.head(10)
```

```python
#MAE abstract error yani mutlak değgerini alıp hatayı çözümleme
df_hata['abstract error'] = np.abs(df_hata['error'])
```

```python
df_hata.head(10)
```

```python
df_hata['percent error'] = np.abs((y-y_tahmin)/y)*100
```

```python
df_hata.head(10)
```

```python
df_hata.mean()
```
