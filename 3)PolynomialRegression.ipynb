```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model  import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("3-customersatisfaction.csv")
```

```python
df.head()
```

```python
df = df.drop("Unnamed: 0", axis=1)
```

```python
df
```

```python
df.info()
```

```python
plt.scatter(df["Customer Satisfaction"], df["Incentive"],color="b")
plt.xlabel("Customer Satisfaction")
plt.ylabel("Incentive")
plt.show()    #polinomal şekilde bir eğri görüyoruz. Bu yüzden polinom regresyon  kullanılabilir.
```

```python
#dependent - independent features
X = df[["Customer Satisfaction"]]
y = df["Incentive"]
```

```python
X
```

```python
y
```

```python
#train-test-split
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
X_train
```

```python
X_test.shape
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
score = r2_score(y_test, y_pred)
print(score)
```

```python
from sklearn.model_selection import cross_val_score


scores = cross_val_score(model, X, y, cv=5, scoring='r2')
print(f"Ortalama R2: {np.mean(scores):.2f}, Standart sapma: {np.std(scores):.2f}")
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train),color="r")
plt.show()
```

```python
poly = PolynomialFeatures(degree=2, include_bias=True)
```

```python
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
y_pred = model.predict(X_test_poly)
score = r2_score(y_test, y_pred)
print(score)
```

```python
def poly_regression(degree):
    poly_features = PolynomialFeatures(degree=degree)
    model = LinearRegression()
    scaler = StandardScaler()
    pipeline = Pipeline([
        ("scaler", scaler),
        ("poly features", poly_features),
        ("model", model),
    ]
    )
    pipeline.fit(X_train, y_train)
    score= pipeline.score(X_test, y_test)
    print("R2 score" ,score)

    y_pred_new = pipeline.predict(X_new)
    plt.plot(X_new, y_pred_new, "r", label="New Predictions")
    plt.scatter(X_train, y_train,  label="Training Points")
    plt.scatter(X_test, y_test,  label="Test Points")
    plt.legend()
    plt.show()
```

```python
for degree in [1,2,4,5,6,7,8,9,10]:
    poly_regression(degree)
```
