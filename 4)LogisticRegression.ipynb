```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = ["subscribed"])
y = df["subscribed"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ",score)
print(classification_report(y_pred,y_test)) 
print("confusion matrix: \n",confusion_matrix(y_pred,y_test))
```

```python
#hyperparameter tuning 
```

```python
model2 = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solver_type = ["lbfgs","liblinear","saga","newton-cg","sag","newton-cholesky"]
max_iter_value = [10,100,500,1000]
```

```python
params =dict(penalty=penalty, C=c_values, solver=solver_type, max_iter=max_iter_value)
```

```python
params
```

```python
#grid search cv -- verdiğimiz parametrelerş kombinasyonluyor ve en iyi sonucu verenleri bize gösteriyor
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model2, param_grid=params, cv=cv, scoring="accuracy", n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_  # en iyi parametrelerimiz bunlarmış.
```

```python
grid.best_score_  
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ",score)
print(classification_report(y_pred,y_test)) 
print("confusion matrix: \n",confusion_matrix(y_pred,y_test))
```
